```python
def bubble_sort(list):
    for i in range(len(list)-1, 0, -1):
        for j in range(i):
            if(list[j] > list[j+1]):
                temp=list[j]
                list[j]=list[j+1]
                list[j+1]=temp
    
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        num = arr[i]
        j = i - 1
        while j >= 0 and num < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = num
```

```python
def swap(list, i, j):
    list[i],list[j] = list[j], list[i]
```

```python
def siftDown(list, i, upper):
    while (True):
        l, r=i*2+1, i*2+2 #childern of any parent node
        #check if there two childern, one child or no children

        if max(i,r) < upper:
            if list[i] >= max(list[l], list[r]): break#check if parent is greater than it childern
            elif list[l] > list[r]:
                swap(list, i, l)
                #update the parent
                i = l
            else:
                swap(list, i, r)  
                i=r  
        #If only one child exist
        elif l < upper:
            if list[l] > list[i]:
                swap(list, i, l)
                i = l
            else: break

        elif r < upper:
            if list[r] > list[i]:
                swap(list, i, r) 
                i=r
            else: break
        else: break        
```

```python
def heap_sort(lst):
    #first step is to heapfy to a max_heap the second step is to sort
    #len(list-2)//2 is the last parent
    #end at -1(last element), and decrement by 1
    for j in range(len(lst)-2//2, -1, -1):
        siftDown(lst, j, len(lst))

    for end in range(len(lst)-1, 0, -1):
        swap(lst, 0, end)
        siftDown(lst, 0, end)
```

```python
def binary_search(arr, num):
    low = 0
    high = len(arr)-1
    
    while low <= high:
        mid=(low+high)//2

        if arr[mid] == num:
            globals()['position'] = mid
            return True
            
        else:
            if arr[mid] < num:
                low = mid

            else:
                high = mid
                
```
